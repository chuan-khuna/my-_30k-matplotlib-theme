```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from utils.vis_utils import *
from utils.vis_utils.palettes import ft, economist, nzk, london_dark
```

```python
n = 400
num_series = 5

df = pd.DataFrame(
    np.random.randn(n, num_series),
    index=np.arange(n),
    columns=[f'fake {i+1}' for i in range(num_series)],
)

df = df.cumsum()
```

```python
use_style('ft')
set_seaborn_palette(ft.vis_colors.values())

fig = plt.figure(figsize=(8, 4))

for col in df.columns:
    sns.lineplot(
        x=df.index, y=df[col], path_effects=get_path_effects(3, foreground=ft.text_colors['wheat'])
    )
```

```python
use_style('nzk')
set_seaborn_palette(nzk.main_colors.values())

fig = plt.figure(figsize=(8, 4))
for col in df.columns:
    sns.lineplot(
        x=df.index, y=df[col], path_effects=get_path_effects(3, foreground=nzk.text_colors['bg'])
    )
```

```python
use_style('economist')
set_seaborn_palette(economist.main_colors.values())

fig = plt.figure(figsize=(8, 4))
for col in df.columns:
    sns.lineplot(
        x=df.index,
        y=df[col],
        path_effects=get_path_effects(3, foreground=economist.text_colors['bg']),
    )
```

```python
use_style('london_dark')
set_seaborn_palette(london_dark.main_colors.values())

fig = plt.figure(figsize=(8, 4))
for col in df.columns:
    sns.lineplot(
        x=df.index,
        y=df[col],
        path_effects=get_path_effects(3, foreground=london_dark.text_colors['bg-light']),
    )
```
